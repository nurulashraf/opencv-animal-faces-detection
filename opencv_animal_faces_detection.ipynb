# Version 1 (Simple)

```python
import cv2
from google.colab.patches import cv2_imshow

image = cv2.imread('/content/cat.jpg')
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Load the pre-trained Haar Cascade Classifier for face detection
face_detection = cv2.CascadeClassifier('/content/haarcascade_frontalcatface.xml')
# cv2.CascadeClassifier() -> This is a function in OpenCV that creates a cascade classifier object.
# cv2.data.haarcascade -> Provides the file path to the directory where OpenCV stores its pre train Haar Cascade classifier files.
# 'haarcascade_frontalface_default.xml' This is a name of a specific XML file that contains the data for the pre-trained Haar Cascasde classifier for detecting frontal human face

# Detects the face in the image
faces = face_detection.detectMultiScale(gray_image, scaleFactor=1.1, minNeighbors=5, minSize = (30,30))
# Face detection object that searches for the target object (faces) in the input images
# Scale Factor = 1.1 -> This parameter specifies how much the image size is reduced at each image scale.
# By scaling down the image the detector can find face of different size. 1.1 represent the image is reduce by 10% at each step.
# minNeighbours = 5 - > This parameter specify how many neighbours each candidate rectangle should have to be considered as a valid detection.
# minSize = (30,30) defines the minimum possible size of the object to be detected. Objects smaller than this size can be ignored.

# Draw the rectangle to inform the detection
for (x, y, w, h) in faces:
  cv2.rectangle(image, (x,y), (x+w,y+h), (0,255,0), 3)

# (x,y) -> These are the coordinates of the top-left corner of the rectangle
# (x+w,y+h) -> These are coordinates of the bottom right corner of the rectangle
# Color -> BGR (B, G, R)

cv2_imshow(image)
```

# Version 2 (Detailed)

## Import Libraries

```python
import cv2
from google.colab.patches import cv2_imshow
```

## Load Haar Cascade Classifiers

```python
face_detection = cv2.CascadeClassifier('/content/haarcascade_frontalcatface.xml')
```

## Function for Face Detection

```python
def detect_cats(image):
  # Convert BGR to RGB
  cat_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
  # Detects the face in the image
  cat_rectangle = face_detection.detectMultiScale(cat_rgb, scaleFactor=1.1, minNeighbors=5, minSize = (30,30))

  # Draw the rectangle to inform the detection
  for (x, y, w, h) in cat_rectangle:
    cv2.rectangle(cat_rgb, (x,y), (x+w,y+h), (0,255,0), 3)

  return cat_rgb
```

## Display Image with Detected Face

```python
image = cv2.imread('/content/cat.jpg')

detect_cats(image)
```
